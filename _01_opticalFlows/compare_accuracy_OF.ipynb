## FB and Warp to define different OFs precision

```python
# %%
# Environment & imports
%matplotlib inline
import cv2
import os
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
from tqdm import tqdm

print('OpenCV version:', cv2.__version__)

# -------------------------------------------------------------
# 1. CONFIGURATION & PATHS
# -------------------------------------------------------------

# Base directory containing "blasto" and "no_blasto" subfolders
BASE_IMG_DIR = Path('/home/phd2/Scrivania/CorsoData/blastocisti')
CSV_PATH = Path("/home/phd2/Scrivania/CorsoRepo/cellPIV/datasets/Farneback/FinalDataset.csv")
OUTPUT_COMPARISON_PATH = Path("/home/phd2/Scrivania/CorsoRepo/cellPIV/_01_opticalFlows/compare_accuracy_OF_results")
OUTPUT_COMPARISON_PATH.mkdir(parents=True, exist_ok=True)

# Processing Options
N_RANDOM_VIDEOS = 100
MAX_FRAMES_PER_VIDEO = 50       # Analyze first 50 frames per video
FRAME_STEP = 1
RANDOM_SEED = 42

# -------------------------------------------------------------
# 2. PARAMETERS SETUP
# -------------------------------------------------------------

# Farneback Params (Dense)
PARAMS_FARNEBACK = {
    'pyr_scale': 0.5,
    'levels': 4,
    'winsize': 25,
    'iterations': 3,
    'poly_n': 5,
    'poly_sigma': 1.2,
    'flags': 0
}

# Lucas-Kanade Params (Sparse)
PARAMS_LK = {
    'winSize': (25, 25),
    'maxLevel': 4,
    'criteria': (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 0.03)
}
# Good Features to Track Params (for LK initialization)
PARAMS_GFTT = {
    'maxCorners': 300,
    'qualityLevel': 0.05,
    'minDistance': 5,
    'blockSize': 7
}

# -------------------------------------------------------------
# 3. UTILITIES
# -------------------------------------------------------------

def get_random_video_ids(csv_path: Path, n=10, seed=42):
    """Reads CSV and returns N random video IDs (dish_well)."""
    if not csv_path.exists():
        raise FileNotFoundError(f"CSV not found at {csv_path}")
    
    df = pd.read_csv(csv_path)
    if 'dish_well' not in df.columns:
        raise ValueError("Column 'dish_well' not found in CSV")
        
    unique_ids = df['dish_well'].unique().tolist()
    # Remove any empty/nan strings if present
    unique_ids = [str(x).strip() for x in unique_ids if str(x).strip() != 'nan']
    
    random.seed(seed)
    if len(unique_ids) > n:
        selected = random.sample(unique_ids, n)
    else:
        selected = unique_ids
    
    return selected

def find_video_path(base_dir: Path, video_id: str):
    """Searches for video_id folder inside 'blasto' or 'no_blasto' subdirs."""
    # Check in 'blasto'
    p1 = base_dir / 'blasto' / video_id
    if p1.exists() and p1.is_dir():
        return p1
    # Check in 'no_blasto'
    p2 = base_dir / 'no_blasto' / video_id
    if p2.exists() and p2.is_dir():
        return p2
    return None

def read_frames(input_path: Path, max_frames: int=None, step: int=1, random_start_frame: bool=False):
    """Reads sorted jpg frames from directory."""
    frames_gray = []
    if not input_path.exists():
        return []

    # 1. Get all files sorted first
    all_files = sorted([p for p in input_path.iterdir() if p.suffix.lower() in ('.jpg', '.jpeg', '.png')])
    
    if len(all_files) == 0:
        return []

    # 2. Determine Start Index
    start_idx = 0
    if random_start_frame:
        # Ensure we don't pick an index out of bounds
        start_idx = random.randint(0, max(0, len(all_files) - 1))

    # 3. Apply Slicing (Step is handled here ONCE)
    files_to_process = all_files[start_idx::step]

    # 4. Read Loop
    for p in files_to_process:
        # Read as grayscale directly
        frame = cv2.imread(str(p), cv2.IMREAD_GRAYSCALE)
        if frame is None: continue
        
        frames_gray.append(frame)
        
        # Stop if we have enough frames
        if max_frames and len(frames_gray) >= max_frames: 
            break
            
    return frames_gray

# --- Metrics Functions (Fixed) ---

def compute_farneback_metrics(prev_gray, next_gray, params):
    # Forward
    flow_fwd = cv2.calcOpticalFlowFarneback(prev_gray, next_gray, None,
                                            params['pyr_scale'], params['levels'], params['winsize'],
                                            params['iterations'], params['poly_n'], params['poly_sigma'],
                                            params.get('flags', 0))
    # Backward
    flow_bwd = cv2.calcOpticalFlowFarneback(next_gray, prev_gray, None,
                                            params['pyr_scale'], params['levels'], params['winsize'],
                                            params['iterations'], params['poly_n'], params['poly_sigma'],
                                            params.get('flags', 0))
    
    # FB Error
    h, w = flow_fwd.shape[:2]
    
    # FIX: Use meshgrid instead of manual tile/arange
    # Create grid of coordinates
    grid_x, grid_y = np.meshgrid(np.arange(w), np.arange(h))
    
    # Add forward flow to grid to get the mapping to next frame
    map_x = (grid_x + flow_fwd[..., 0]).astype(np.float32)
    map_y = (grid_y + flow_fwd[..., 1]).astype(np.float32)
    
    # Remap the backward flow to the current frame coordinates
    warped_bwd_x = cv2.remap(flow_bwd[..., 0], map_x, map_y, cv2.INTER_LINEAR)
    warped_bwd_y = cv2.remap(flow_bwd[..., 1], map_x, map_y, cv2.INTER_LINEAR)
    
    # Consistency check: Forward + Warped_Backward should be 0
    diff_x = flow_fwd[..., 0] + warped_bwd_x
    diff_y = flow_fwd[..., 1] + warped_bwd_y
    fb_error = np.mean(np.sqrt(diff_x**2 + diff_y**2))

    # Photo Error
    # Remap previous image using flow to match next image
    warped_prev = cv2.remap(prev_gray, map_x, map_y, cv2.INTER_LINEAR)
    
    # Calculate difference, ignoring invalid borders (where remap puts 0)
    # Simple absolute difference on valid pixels
    err_map = np.abs(warped_prev.astype(np.float32) - next_gray.astype(np.float32))
    photo_error = np.mean(err_map)

    return fb_error, photo_error

def compute_lk_metrics(prev_gray, next_gray, lk_params, gftt_params):
    p0 = cv2.goodFeaturesToTrack(prev_gray, mask=None, **gftt_params)
    if p0 is None or len(p0) == 0: return np.nan, np.nan

    p1, st, err = cv2.calcOpticalFlowPyrLK(prev_gray, next_gray, p0, None, **lk_params)
    
    # Keep only good points
    if p1 is None: return np.nan, np.nan
    good_p1 = p1[st == 1]
    good_p0 = p0[st == 1]
    
    if len(good_p1) == 0: return np.nan, np.nan

    # Backward tracking
    p0r, st_inv, err_inv = cv2.calcOpticalFlowPyrLK(next_gray, prev_gray, good_p1, None, **lk_params)
    
    # FB Error (Euclidean dist)
    dist = np.sqrt(np.sum((good_p0 - p0r)**2, axis=1))
    fb_error = np.mean(dist)

    # Photo Error (Sparse sampling)
    def sample_points(img, points):
        samples = []
        for pt in points:
            # Safe subpixel access
            if 0 <= pt[1] < img.shape[0]-1 and 0 <= pt[0] < img.shape[1]-1:
                val = cv2.getRectSubPix(img.astype(np.float32), (1,1), tuple(pt))
                samples.append(val[0][0])
            else:
                samples.append(0)
        return np.array(samples)

    vals_prev = sample_points(prev_gray, good_p0)
    vals_next = sample_points(next_gray, good_p1)
    
    if len(vals_prev) == 0: return np.nan, np.nan
    photo_error = np.mean(np.abs(vals_prev - vals_next))

    return fb_error, photo_error

# -------------------------------------------------------------
# 4. MAIN EXECUTION LOOP
# -------------------------------------------------------------

# 1. Select Videos
target_videos = get_random_video_ids(CSV_PATH, n=N_RANDOM_VIDEOS, seed=RANDOM_SEED)
print(f"Selected {len(target_videos)} videos for analysis: {target_videos}")

# Storage for aggregated results
all_metrics = []

for vid_idx, vid_id in enumerate(target_videos):
    # 2. Find Path
    vid_path = find_video_path(BASE_IMG_DIR, vid_id)
    if vid_path is None:
        print(f"[{vid_idx+1}/{N_RANDOM_VIDEOS}] Skipping {vid_id} (Folder not found in blasto/no_blasto)")
        continue
        
    print(f"[{vid_idx+1}/{N_RANDOM_VIDEOS}] Processing {vid_id}...")
    
    # 3. Read Frames
    frames = read_frames(vid_path, max_frames=MAX_FRAMES_PER_VIDEO, step=FRAME_STEP, random_start_frame=True)
    if len(frames) < 2:
        print(f"  > Not enough frames ({len(frames)}). Skipping.")
        continue

    # 4. Compute Metrics per frame
    for i in tqdm(range(len(frames) - 1), desc=f"  > Frames", leave=False):
        prev_g = frames[i]
        next_g = frames[i+1]
        
        try:
            # Farneback
            fb_fb_err, fb_ph_err = compute_farneback_metrics(prev_g, next_g, PARAMS_FARNEBACK)
            
            # LK
            lk_fb_err, lk_ph_err = compute_lk_metrics(prev_g, next_g, PARAMS_LK, PARAMS_GFTT)
            
            all_metrics.append({
                'video': vid_id,
                'frame_idx': i,
                'Method': 'Farneback',
                'FB_Error': fb_fb_err,
                'Photo_Error': fb_ph_err
            })
            all_metrics.append({
                'video': vid_id,
                'frame_idx': i,
                'Method': 'Lucas-Kanade',
                'FB_Error': lk_fb_err,
                'Photo_Error': lk_ph_err
            })
        except Exception as e:
            print(f"Error on frame {i}: {e}")
            continue

# -------------------------------------------------------------
# 5. AGGREGATE RESULTS & PLOT
# -------------------------------------------------------------

df_all = pd.DataFrame(all_metrics)

if df_all.empty:
    print("No metrics computed. Check paths.")
else:
    # Summary Table
    summary = df_all.groupby('Method')[['FB_Error', 'Photo_Error']].agg(['mean', 'std'])
    print("\n=== FINAL SUMMARY (Average over videos) ===")
    print(summary)
    
    

    # Plot Distribution
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    
    # Boxplot for FB Error
    df_clean = df_all.dropna()
    
    if not df_clean.empty:
        df_clean.boxplot(column='FB_Error', by='Method', ax=axes[0], grid=True)
        axes[0].set_title('Forward-Backward Consistency Error distribution')
        axes[0].set_ylabel('Error (Pixels)')
        axes[0].set_xlabel('')
        
        # Boxplot for Photo Error
        df_clean.boxplot(column='Photo_Error', by='Method', ax=axes[1], grid=True)
        axes[1].set_title('Photometric Warp Error distribution')
        axes[1].set_ylabel('Intensity Difference (0-255)')
        axes[1].set_xlabel('')
        
        plt.suptitle(f'Comparison on {len(target_videos)} Random Videos', fontsize=16)
        plt.tight_layout()
        plt.show()
        
        # Save CSV
        out_csv = OUTPUT_COMPARISON_PATH / f"optical_flow_comparison_{N_RANDOM_VIDEOS}videos.csv"
        df_all.to_csv(out_csv, index=False)
        print(f"Detailed metrics saved to {out_csv}")
    else:
        print("No valid data to plot.")
```
